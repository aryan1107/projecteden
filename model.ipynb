```python
from tensorflow import keras
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler    
from sklearn.metrics import confusion_matrix
```

```python
# loading data
crop_data = pd.read_csv('/Users/aryan/Desktop/ResearchClass/ProjectEden/Crop_recommendation.csv')
crop_data.head()
```

```python
# loading labels
crop_old = crop_data['label'].unique()
crop_old
```

```python
# replace values using map basically converting it to numbers to train model
crop_data['label'] = crop_data['label'].map({'rice': 0, 'maize': 1, 'chickpea': 2, 'kidneybeans': 3, 'pigeonpeas': 4, 
                    'mothbeans': 5, 'mungbean': 6, 'blackgram':7, 'lentil': 8, 'pomegranate': 9,
                    'banana': 10, 'mango': 11, 'grapes': 12, 'watermelon': 13, 'muskmelon': 14, 'apple':15,
                    'orange': 16, 'papaya':17, 'coconut':18, 'cotton':19, 'jute': 20, 'coffee': 21})
crop_data.head()
```

```python
# Target is data we want to predict in this case that is plant class
y = crop_data.label

#  columns we need to drop that is not needed
drop_cols = ['label']

# feature matrix below since droping cols we need to set axis to 1 if it was rows then 0
x = crop_data.drop(drop_cols, axis = 1)
x.head()
```

```python
# Split Train and Test Datasets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25) 

# normalize the dataset i.e. X_train and X_test
scaler = MinMaxScaler(feature_range=(0, 1))
X_train = scaler.fit_transform(X_train)    
X_test = scaler.fit_transform(X_test)
```

```python
# neuron network model
def flower_model():
    model = tf.keras.models.Sequential([ 
      tf.keras.layers.Dense(14, activation='relu'),
      tf.keras.layers.Dense(7, activation='relu'),
      tf.keras.layers.Dense(22, activation='softmax')
    ]) 

    ### END CODE HERE

    # Compile the model
    model.compile(optimizer='adam', 
                  loss='sparse_categorical_crossentropy', 
                  metrics=['accuracy']) 


    return model
```

```python
# Save your untrained model
model = flower_model()

#training model
train = model.fit(X_train, y_train, epochs=100, batch_size=32, verbose = 1)
```

```python
new_y = [0.065862, 0.065862 , 0.857653]
prediction = model.predict([new_y])[0]
print(prediction)
```

```python
# evaluating on unseen data our accuracy on unseen data is around accuracy: 90.18%
scores = model.evaluate(X_test, y_test, batch_size=32)
print(); print(model.metrics_names); print(scores)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
